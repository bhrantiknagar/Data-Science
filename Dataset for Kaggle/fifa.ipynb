```python
data = {
    "HomeGoals": [4, 3, 2, 3, 1],
    "AwayGoals": [1, 0, 1, 1, 0],
    "Attendance": [4444, 18346, 24059, 2549, 23409],
    "Winner": ["France", "USA", "Yugoslavia", "Romania", "Argentina"]
}

df = pd.DataFrame(data)

x = df[["HomeGoals", "AwayGoals", "Attendance"]]
y = df["Winner"]

model = DecisionTreeClassifier()
model.fit(x, y)

home_goals = int(input("Enter Home Team Goals: "))
away_goals = int(input("Enter Away Team Goals: "))
attendance = int(input("Enter Attendance: "))

new_data = pd.DataFrame({
    "HomeGoals": [home_goals],
    "AwayGoals": [away_goals],
    "Attendance": [attendance]
})

prediction = model.predict(new_data)
print("\nPredicted Winner:", prediction[0])

y_pred = model.predict(x)
accuracy = accuracy_score(y, y_pred)

print(f"\nAccuracy: {accuracy * 100:.2f}%")
```
