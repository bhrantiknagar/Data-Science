```python
df = {
    "bedrooms":data["bedrooms"],
    "yr_built":data["yr_built"],  
    "bathrooms":data["bathrooms"],  
    "floors":data["floors"],  
    "price":data["price"]
}
```

```python
from sklearn.linear_model import LinearRegression

# Dataset
da = pd.DataFrame(df)

# Features and Target
x = da[["bedrooms", "yr_built", "bathrooms", "floors"]]
y = da["price"]

# Model
model = LinearRegression()

# Train
model.fit(x, y)

# Model Details
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# User Input
bedrooms = int(input("Enter number of bedrooms: "))
year = int(input("Enter year built: "))
bathroom = float(input("Enter number of bathrooms: "))
floors = int(input("Enter number of floors: "))

# Prediction
price = model.predict([[bedrooms, year, bathroom, floors]])

print(f"\nPredicted Price = ₹{price[0]:,.2f}")

# Predicted values
predicted_price = model.predict(x)

# Actual vs Predicted Plot
plt.figure(figsize=(8,5))
plt.scatter(y, predicted_price)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.grid(True)
plt.show()
```
