```python
df = {
    "bedrooms":data["bedrooms"],
    "yr_built":data["yr_built"],  
    "bathrooms":data["bathrooms"],  
    "floors":data["floors"],  
    "price":data["price"]
}
```

```python
# Dataset

da=pd.DataFrame(df)


x = da[["bedrooms","yr_built","bathrooms","floors"]]
y = da["price"]


model=LinearRegression()


model.fit(x,y)


print("Coefficients:", model.coef_[0])
print("Intercept:", model.intercept_)


bedrooms = float(input("Enter no. of bedroom: "))
bathroom = float(input("Enter no. of bathroom: "))
floors = int(input("Enter no. of floor : "))
year = int(input("Enter no. of year : "))


price = model.predict([[bedrooms, floors, year, bathroom]])


print(f"\nPredicted Price = ₹{price[0]:.2f}")

plt.figure(figsize=(8,5))
plt.scatter(da["bedrooms"], df["price"])
salary_pred = model.predict(x)
plt.plot(df["price"], salary_pred)
plt.xlabel("Bedroom ")
plt.ylabel("Price")
# plt.title("Multiple Linear Regression")
plt.grid()
plt.legend()
plt.show()
```
