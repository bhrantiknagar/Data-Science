```python

zf=(np.random.rand(4,4)*100).astype(int)
print(zf)
zf[2,:]                               # zf[row,col]
zf[0,:]
zf[:,3]
zf[:,3]
```

```python
lu=(np.random.rand(5,5)*100).astype(int)
print(lu)
lu[:]                           # full
lu[0]                           # first row
lu[-1]                          # last row
lu[:3]                          # first 3 row
lu[-3:]                         # last 3 row
lu[2:6]                         # 2 row to 5 row
lu[:4]                          # 4 index row
lu[::2]                         # every 2nd row
lu[::3]                         # every 3rd row
lu[::-1]                        # reverse array
lu[::-2]                        # reverse every 2nd array
```

```python
le=(np.random.rand(5,5)*100).astype(int)
print(le)
le[2,-1]
le[:,-3]
```

```python
no=(np.random.rand(3,3)*100).astype(int)
print(no)
no[0]                          # first row
no[1]                          # second row
no[2]                          # third row
no[-1]                         # last row
no[0:2]                        # first two row
no[1:]                         # last two row
no[0:3]                        # multiple row
no[1,:]                        # select row 1
no[2,:]                        # select row 2
no[:,:]                        # all rows
```
