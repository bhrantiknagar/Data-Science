```python

l1=[10,20,30,40,50]
l2=[60,70,80,90,100]                      # in list
print("length: ", len(l1))
print("max: ", max(l1))
print("min: ", min(l1))
print("sum: ", sum(l1))
print("sorted: ", min(l1))
for i in reversed(l1):
  print(i,end="")
print()
print("enumerate: ", end=" ")
for i in enumerate(l1):
  print(i,end=" ")
print()
print("zip: ", end=" ")
for i in zip(l1,l2):
  print(i, end=" ")
```
