Logistic Regression :-

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score 
```

Result Status Model :-

```python
data = {
    "Marks":[35,45,55,65,75],
    "StudyHours":[1,2,3,4,5],
    "Attendace":[50,60,70,80,90],
    "Result":["Fail","Fail","Pass","Pass","Pass"]
}

df = pd.DataFrame(data)
print(df)

x = df[["Marks", "StudyHours", "Attendace"]]
y = df["Result"]

model = LogisticRegression()

model.fit(x,y)

marks = float(input("Enter Marks: "))
hours = float(input("Enter Study Hours: "))
attendance = float(input("Enter Attendance: "))

prediction = model.predict([[marks, hours, attendance]])
print("Prediction:", prediction[0])

probability = model.predict_proba([[marks, hours, attendance]])

print("\nProbability:")
print("Fail Probability =", round(probability[0][0]* 100, 2), "%")
print("Pass Probability =", round(probability[0][1]* 100, 2), "%")

y_pred = model.predict(x)
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy * 100, "%")
```

Loan Apporval / Rejection Model :-

```python
data = {
    "income": [20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000],
    "creditscore": [500, 550, 580, 600, 620, 650, 680, 700, 720, 750],
    "loan_amount": [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000],
    "status": ["Rejected","Rejected","Rejected","Rejected","Approved","Approved","Approved","Approved","Approved","Approved"]
}

df = pd.DataFrame(data)
print(df)

x = df[["income", "creditscore", "loan_amount"]]
y = df["status"]

model = LogisticRegression()

model.fit(x,y)

income = float(input("Enter Income: "))
creditscore = float(input("Enter Credit Score: "))
loan_amount = float(input("Enter Loan Amount: "))

prediction = model.predict([[income, creditscore, loan_amount]])
print("Prediction:", prediction[0])

probability = model.predict_proba([[income, creditscore, loan_amount]])

print("\nProbability:")
print("Reject Probability =", round(probability[0][0]* 100, 2), "%")
print("Approve Probability =", round(probability[0][1]* 100, 2), "%")

y_pred = model.predict(x)
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy * 100, "%")
```

Covid-19 Check-up Model :-

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

data = {
    "Fever":[98,99,100,101,102,103,98,99,101,102],
    "Cough":[0,0,1,1,1,1,0,1,1,1],
    "BreathingProblem":[0,0,0,1,1,1,0,0,1,1],
    "Covid_Status":[
        "Nagative", "Nagative","Nagative",
        "Positive","Positive","Positive",
        "Nagative","Nagative","Positive","Positive"
    ]
}

df = pd.DataFrame(data)

x = df[["Fever", "Cough", "BreathingProblem"]]
y = df["Covid_Status"]

model = DecisionTreeClassifier()

model.fit(x,y)

fever = float(input("Enter Temperature (°F): "))
cough = int(input("Cough? (1=Yes, 0=No):  "))
breathing = int(input("Breathing Problem? (1=Yes, 0=No): "))

prediction = model.predict([[fever, cough, breathing]])
print("\nCovid Status:", prediction[0])

y_pred = model.predict(x)
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy * 100, "%")
```
