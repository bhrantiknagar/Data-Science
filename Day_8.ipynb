MatPlotLib :-

```python
import matplotlib.pyplot as lab
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

lab.plot(x, y)
lab.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
lab.plot(x, y)
lab.grid(True)
lab.show()
# grid(True) -> enable grid
```

```python
students = ["A", "B", "C", "D"]
marks = [40, 20, 40, 20]
lab.bar(students, marks)
colors = ['lightgray', 'darkgray', 'gray', 'black']
lab.bar(students, marks, color=colors)
lab.show()

# bar() -> for bar graph
```

```python
students = ["A", "B", "C", "D"]
marks = [40, 20, 40, 20]
lab.barh(students, marks)
colors = ['lightgray', 'darkgray', 'gray', 'black']
lab.bar(students, marks, color=colors)
lab.show()
```

```python
import matplotlib.pyplot as lab

x = [1,2,3,4]
y = [5,10,15,20]
lab.plot(x, y,linestyle="-",linewidth=4)
lab.title("Simple Line Graph")
lab.xlabel("X-axis")
lab.ylabel("Y-axis")
lab.show()
```
