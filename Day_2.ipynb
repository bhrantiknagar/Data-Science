```python
# Questions:-

emp=2345
salary=40000.5
emp_name="lucky"
emp_status=True

print("\n")

print("type:-")
print(type(emp))
print(type(salary))
print(type(emp_name))
print(type(emp_status))

print("\n")

print("output:-")
print(emp)
print(salary)
print(emp_name)
print(emp_status)

print("\n")

print("id:-")
print(id(emp))
print(id(salary))
print(id(emp_name))
print(id(emp_status))
```

```python
# Assignment opereators:-


# Bitwise operator:-

# bitwise AND
print(2&3)

# bitwise OR
print(5|4)
# Exclusive XOR
print(3^2)

# bitwise NOT
print(~5)

# bitwise RIGHTSHIFT
print(12>>2)
print(90>>5)

# bitwise LEFTSHIFT
print(13<<6)
print(9<<3)
```

```python
# Membership Operators:-

# in
# in not

a=[10,20,30,40,50]
print(100 in a)

```

```python
# Identity Operators:-

# is
# is not

a=257
b=a
c=257
print(a is b)
print(a is c)
print(a is not c)
```
