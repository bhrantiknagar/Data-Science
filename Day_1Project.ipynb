```python
# Create a python program that asks for a student's name and marks, calculates the total and percentage, and displays the result.

print("="*40)
print("               Marksheet            ")
print("="*40)
name=input("Enter your name:")
m1=int(input("Enter marks of first subject:"))
m2=int(input("Enter marks of second subject:"))

total=m1+m2
percentage=(total/200)*100

print("\nTotal number:", total)
print("percentage:", percentage)

print("\n")
print("="*40)
print("                RESULT               ")
print("="*40)

print("Name\t\t : \t",name)
print("Total Marks\t : \t",total)
print("percentage\t : \t",percentage)
print("\n")
```
