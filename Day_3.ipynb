```python
for i in range(1,7):
  for j in range(i):
    j+=1
    print(j, end=" ")
  print("")
```

```python

i=1
while i<=2:
  j=1
  while j<=2:
    k=1
    while k<=2:
      l=1
      while l<=2:
        print(i,j,k,l)
        l+=1
      k+=1
    j+=1
  i+=1
```

```python
for i in range(1,7):
  for j in range(1,3):
    j+=1
    print("*", end="")
  i+=1
  print(" ")
```

```python

row=int(input("Enter a row:"))
col=int(input("Enter a col:"))
print("\n")

for i in range(row):
  for j in range(col):
    j+=1
    print("^", end="")
  i+=1
  print("")
```
