```python
# ==========================================
#         CRICKET SCORE ANALYZER
# ==========================================

print("=" * 45)
print("         CRICKET SCORE ANALYZER")
print("=" * 45)

overs = int(input("Enter Number of Overs Played : "))

total_runs = 0
fours = 0
sixes = 0
dot_balls = 0

# Nested While Loop for Overs and Balls

over = 1
while over <= overs:
    print(f"\n---------- Over {over} ----------")

    ball = 1
    while ball <= 6:
        runs = int(input(f"Ball {ball} Runs (0-6) : "))

        # Validation
        if runs < 0 or runs > 6:
            print("Invalid Runs! Counting as Dot Ball.")
            runs = 0

        # Update Score
        total_runs += runs

        # Boundary & Dot Ball Count
        if runs == 4:
            fours += 1
        elif runs == 6:
            sixes += 1
        elif runs == 0:
            dot_balls += 1

        ball += 1

    over += 1

# Calculations
total_balls = overs * 6
boundary_runs = (fours * 4) + (sixes * 6)
strike_rate = (total_runs / total_balls) * 100

# Performance Analysis
if strike_rate >= 150:
    performance = "Excellent 🏆"
elif strike_rate >= 100:
    performance = "Good 👍"
else:
    performance = "Needs Improvement 😔"

# Additional Analysis
if total_runs >= 100:
    print("\nCentury Partnership Level Score! 🔥")
elif total_runs >= 50:
    print("\nHalf-Century Level Score! 👏")
else:
    print("\nKeep Practicing! 💪")

# Final Report
print("\n" + "=" * 45)
print("           MATCH SUMMARY")
print("=" * 45)
print(f"Total Overs Played : {overs}")
print(f"Total Balls        : {total_balls}")
print(f"Total Runs         : {total_runs}")
print(f"Fours              : {fours}")
print(f"Sixes              : {sixes}")
print(f"Boundary Runs      : {boundary_runs}")
print(f"Dot Balls          : {dot_balls}")
print(f"Strike Rate        : {strike_rate:.2f}")
print(f"Performance        : {performance}")
```
