```python
# for age
age_median = dc["Age"].median()
# age_mean = dc["Age"].mean()
dc["Age"] = dc["Age"].fillna(age_median)
# dc["Age"] = dc["Age"].fillna(age_mean)
```

```python
# for fare
fare_median = dc["Fare"].median()
dc['Fare'] = dc['Fare'].fillna(fare_median)
```

```python
# for cabin
cabin_mode = dc['Cabin'].mode()[0]
print(cabin_mode)
dc['Cabin'] = dc["Cabin"].fillna(cabin_mode)
```
