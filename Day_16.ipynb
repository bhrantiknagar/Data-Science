```python
from jinja2 import Template

def generate_class(class_name, attrs):
    with open('template.txt') as f:
        tmpl = Template(f.read())
    
    # Render the template with your data
    print(tmpl.render(name=class_name, attributes=attrs))

# Usage
generate_class("User", ["username", "email", "password"])
```
