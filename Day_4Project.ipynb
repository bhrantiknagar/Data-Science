```python

print("="*45)
print("             COLLEGE LIBRARY SYSTEM")
print("="*45)

print("\nLibrary Name     : Dusara college ")
print("Location         : Kota")

books= ["Python Programming", 
        "Data Structures", 
        "DBMS", 
        "Computer Networks", 
        "Operating System"]

print("\nAvailable Books: ")

for i in range(0,len(books)):
    print(f"{i+1}. {books[i]}")

categories = ["Programming", "Database", "Networking", "System"]

print("\nBooks Categories:")
for i in range(0,len(categories)):
    print(f"{categories[i]}")

stock = {
    "Python Programming": 10,
    "Data Structures": 8,
    "DBMS": 6,
    "Computer Networks": 5,
    "Operating System": 4,
}


print("\nBook Stock:")

for book, qty in stock.items():
    print(f"{book} : {qty}")

print("\nChecking Book Availability...")

book_name= input("Enter Book Name: ")

print("Book Available in Library")

if book_name in stock:
    print(f"\n{book_name} is Available in Library ")
else:
    print("\nSorry, Book Not Available.")

print(f"\nTotal Books Titles : {len(books)}")
print(f"Total Categories   : {len(categories)}")
```
