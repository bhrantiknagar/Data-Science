```python
gra = data['price']
plt.hist(gra, bins=30, color='white', edgecolor='red')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
pd.crosstab(data['price'], data['sqft_living'])
```

```python
plt.hist(data["yr_built"], alpha=0.5, bins=100, label="Year Built", edgecolor='black')
plt.xlabel('Year Built')
plt.ylabel('Price')
plt.title('Year Built vs Price')
plt.show()
```

```python
plt.scatter(data['yr_renovated'].value_counts(), data['yr_built'].value_counts())
plt.xlabel('Renovation Year')
plt.ylabel('Year Built')
plt.show()
```
